```
# Go through the tweets to analyse thier sentiment 
for tweet in tweets:
    print(tweet.text)
    
    #Step 4 Perform Sentiment Analysis on Tweets
    analysis = TextBlob(tweet.text)
    print(analysis.sentiment)
    print("")
```

We ger two properties polarity (-1 to 1) and subjectivity ( 0 to 1)

Polarity is how negative or positive a tweet is and Subjectivity is to how much of an opinion it is compared to how factual it is.
